```python
#If you note the records above, we still have some charachters which are not detected. 
#We can't have that, let's remove that

companies['permalink']=companies[['permalink']].applymap(lambda x: x.encode('utf-8').decode('ascii', 'ignore'))
companies['name']=companies.name.str.encode('utf-8').str.decode('ascii', 'ignore')
companies
```

```python
#Cannot use top_9_companies since it of type DataframeGroupBy
#Picking countries from the list
top9 = venture_dataframe[venture_dataframe['country_code'].isin(["USA","CHN","GBR","IND","CAN","FRA","ISR","DEU","JPN"])]
top9
```

```python
top9['primary_sector'] =top9['category_list'].astype(str).apply(lambda x: x.split('|')[0])
```

```python
top9
```

```python
mapping['category_list']=mapping['category_list'].str.replace('0','na')
```

```python
melted=pandas.melt(mapping, id_vars=['category_list'], var_name=['main_sector'])
```

```python
#We need the values which are equal to 1
melted=melted[melted.value==1]
```

```python
melted=melted.drop('value',axis=1)
melted.shape
```

```python
top9
```

```python
top9=pandas.merge(top9,melted,how="left",left_on="primary_sector",right_on="category_list")
```
